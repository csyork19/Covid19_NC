```python
import plotly.figure_factory as ff

import numpy as np
import pandas as pd
```

```python
df_sample = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/live/us-counties.csv')
df_sample_r = df_sample[df_sample['state'] == 'North Carolina']
```

```python
covid_cases = df_sample_r['cases'].tolist()
fips = df_sample_r['fips'].tolist()
```

```python
endpts = list(np.mgrid[min(covid_cases):max(covid_cases):9j])
colorscale = ["#0A2F51",
"#0E4D64",
"#137177",
"#188977",
"#1D9A6C",
"#39A96B",
"#56B870",
"#74C67A",
"#99D492",
"#BFE1B0",
"#DEEDCF"]
```

```python
fig = ff.create_choropleth(
    fips=fips, values=covid_cases, scope=['North Carolina'], show_state_data=True,
    colorscale=colorscale, binning_endpoints=endpts, round_legend_values=True,
    plot_bgcolor='rgb(229,229,229)',
    paper_bgcolor='rgb(229,229,229)',
    legend_title='Covid cases by County',
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5},
    exponent_format=True,
)
```

```python
fig.layout.template = None
fig.show()
```
